```python
import json

RECIPE_DB_PATH = "../data/recipes_en.json" # 전달받은 데이터 파일 경로

#DB에 파일 로드하는 함수
def load_recipe_db():
    try:
        with open(RECIPE_DB_PATH, 'r', encoding='utf-8') as f:
            recipe_db = json.load(f)
        return recipe_db
    except FileNotFoundError:
        print(f"오류: {RECIPE_DB_PATH} 파일을 찾을 수 없습니다.")
        return []
```

```python
# 전역 변수로 레시피 DB를 로드
RECIPES = load_recipe_db()

def search_recipes_v2_substring(translated_ingredients, recipe_db):
    """
    [V2 로직: '부분 일치(포함)' (in)]
    "shrimp"라고 검색하면 "shrimp prawns deveined"를 찾을 수 있도록 구현
    """
    
    for recipe in recipe_db:
        recipe_main_list = recipe.get("main_ingredients", [])
        
        # (설명을 위해 리스트 대신 join/string 검색으로 더 단순화)
        # ["tapioca starch", "garlic powder", "shrimp prawns deveined"]
        # -> "tapioca starch|garlic powder|shrimp prawns deveined"
        # (하나의 긴 문자열로 합쳐서, 그 안에 검색어가 있는지 확인)
        recipe_main_string = "|".join(recipe_main_list).lower()
        
        found_all_ingredients = True
        for search_term in translated_ingredients:
            # (e.g., "shrimp")
            if search_term.lower() not in recipe_main_string:
                found_all_ingredients = False
                break # 이 레시피는 탈락
        
        if found_all_ingredients:
            print(f"[성공] \"{recipe['title']}\"")
            return recipe

    print("[실패] '부분 일치'하는 레시피가 없습니다.")
    return None
```

```python

if __name__ == "__main__":
  
    RECIPES = load_recipe_db()
    
    if RECIPES:
      # T1
        search_1 = ["shrimp", "salt"]
        result_v2 = search_recipes_v2_substring(search_1, RECIPES)
      
      # T2
        search_2 = ["garlic", "fish sauce"]
        result_v2 = search_recipes_v2_substring(search_2, RECIPES)
        
      # T3
        search_3 = ["pork", "garlic"] 
        result_v2 = search_recipes_v2_substring(search_3, RECIPES)
```
